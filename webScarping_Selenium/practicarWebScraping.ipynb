### 1--> Importamos las librerias

```python
import random
from time import sleep
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

### 2--> Nos ubicamos donde se encuentra la información a extraer
<ol>
    <li>Añadimos la ruta del driver correspondiente al navegador desde el cual vamos a realizar las busquedas interactivas[Chrome]</li>
    <li>Indicamos la url de la pagina donde vamos a realizar el web scraping.[Amazon]</li>
    <li>Realizamos un búsqueda y pulsamos el boton para acceder a la sección de la página donde se encuantra la información deseada.[monitor pc]</li>
    <li>Seleccionamos los filtros correspondientes para que la búsqueda sea más adecuada. En este caso el precio[100€-200€]</li>

</ol>

```python
driver = webdriver.Chrome('./chromedriver.exe')
driver.get('https://www.amazon.es/')
buscador = driver.find_element_by_id('twotabsearchtextbox')
buscador.send_keys('monitor pc')
sleep(2.5)

botonBuscador = driver.find_element_by_id('nav-search-submit-button')
botonBuscador.click()
sleep(1.5)

checkbox = driver.find_element_by_xpath("//li[@aria-label='100 - 200 EUR']")
checkbox.click()
sleep(1.5)
```

### 3--> Guardamos la información en listas
<ol>
    <li>Guardamos en una lista los monitores tras seleccionar la parte del código html de la página web</li>
    <li>Recorremos la lista, de cada monitor guardamos el nombre y el precio(Para saber donde se encuentran excactamente el nombre y el precio se ha inspeccionado previamente del código html)</li>
    <li>Finalmente pasamos a la siguiente página</li>
</ol>

```python
listaMonitores = driver.find_elements_by_xpath("//div[@data-component-type='s-search-result']")

nombres = []
precios = []

saveData = list()

for monitor in listaMonitores:
    try:
        #i+=1
        precio = monitor.find_element_by_xpath(".//span[@class='a-price-whole']").text
        nombre = monitor.find_element_by_xpath(".//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']").text
        
        nombres.append(nombre)
        precios.append(precio)
        saveData.append(nombre)
        saveData.append(precio)
        
    except:
        

sleep(3.0)
botonSiguinte = driver.find_element_by_xpath("//li[@class='a-last']//a[1]")
botonSiguinte.click()
    
```

### 4--> Cuadro de mandos con pandas

```python
df = pd.DataFrame(list(zip(nombres,precios)), columns = ['Names','Price'])
df
```

### 5--> Guardamos la información en ficheros (csv y xlxs)

```python
ficheroWS = 'C:/Users/34674/Documents/python/webScraping_Twitter/ficheroWS.csv'
f = open(ficheroWS, 'a+')

for i in saveData:
    f.write(i)
    f.write(';')
f.close()
```
